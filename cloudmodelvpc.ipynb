### 4.2 Algorithm/Pseudocode Used

#### 4.2.1 Multi-layer Perceptron (MLPClassifier) Algorithm

*   **Underlying Algorithm**: The MLPClassifier implements a feedforward artificial neural network trained with backpropagation. The algorithm consists of:
    1.  **Input Layer**: Receives the feature vector of a network flow instance.
    2.  **Hidden Layers**: One or more layers of neurons, each applying a weighted sum of its inputs followed by a non-linear activation function (e.g., `tanh` or `relu`). These layers learn complex patterns and relationships in the data.
    3.  **Output Layer**: Produces the final prediction (e.g., 'threat' or 'non-threat'). For binary classification, often uses a sigmoid activation.
    4.  **Weights and Biases**: Each connection between neurons has an associated weight, and each neuron has a bias. These are the learnable parameters of the network.
    5.  **Forward Propagation**: Input data flows through the network, activating neurons in successive layers, culminating in an output prediction.
    6.  **Backpropagation**: The difference between the predicted output and the true label (the error) is calculated. This error is then propagated backward through the network, and an optimization algorithm (e.g., `Adam` or `SGD`) adjusts the weights and biases to minimize this error over many iterations (epochs).

#### 4.2.2 Decision Tree Classifier Algorithm

*   **Underlying Algorithm**: The DecisionTreeClassifier builds a tree-like model of decisions. It works by recursively partitioning the data space into smaller regions until each region contains instances of predominantly one class. The splitting criteria are chosen to maximize information gain or minimize impurity (e.g., `gini` impurity or `entropy`).
    1.  **Root Node**: Represents the entire dataset.
    2.  **Splitting**: At each node, the algorithm evaluates different features and their thresholds to find the best split that divides the data into subsets that are as homogeneous as possible with respect to the target variable.
    3.  **Decision Nodes**: Represent a test on an attribute (e.g., `bytes_out <= 433.00`).
    4.  **Branches**: Represent the outcomes of the test (e.g., true/false or numerical ranges).
    5.  **Leaf Nodes**: Represent the final class prediction (e.g., `class: 1` or `class: 0`).
    6.  **Pruning (Implicit/Explicit)**: Techniques to reduce the size of the tree to prevent overfitting, often controlled by hyperparameters like `max_depth`, `min_samples_split`, `min_samples_leaf`.

### 4.3 Testing Techniques

Rigorous testing was crucial to ensure the reliability and effectiveness of the threat detection models. Various techniques were employed throughout the project:

*   **Train-Test Split**: The dataset was divided into training (75%) and testing (25%) sets. This fundamental technique ensures that the models are evaluated on unseen data, providing an unbiased estimate of their generalization performance.

*   **Hyperparameter Tuning with Cross-Validation (`GridSearchCV`)**:
    *   **Cross-Validation**: During `GridSearchCV`, the training data was further split into multiple folds (e.g., 5-fold cross-validation). Models were trained on a subset of these folds and validated on the remaining fold, rotating through all possible combinations. This technique provides a more robust estimate of model performance and helps prevent overfitting to a single validation set.
    *   **Grid Search**: This systematically works through multiple combinations of parameter tunes, evaluating each combination's performance using cross-validation. This allowed for the optimization of MLP and Decision Tree hyperparameters to achieve the best F1-score.

*   **Model Evaluation Metrics**: A comprehensive set of metrics was used to assess model performance, each providing different insights:
    *   **Accuracy**: The proportion of correctly classified instances. While easy to understand, it can be misleading in imbalanced datasets.
    *   **Precision**: The proportion of positive identifications that were actually correct. Important for minimizing false positives (e.g., not incorrectly flagging legitimate traffic as a threat).
    *   **Recall (Sensitivity)**: The proportion of actual positives that were correctly identified. Crucial for minimizing false negatives (e.g., not missing actual threats).
    *   **F1-Score**: The harmonic mean of Precision and Recall, providing a balance between the two. Particularly useful in imbalanced classification problems where a high F1-score indicates good performance on the positive class.
    *   **Confusion Matrix**: A table showing the counts of True Positives, True Negatives, False Positives, and False Negatives. This provides a detailed breakdown of classification errors and is invaluable for understanding model behavior, especially in imbalanced scenarios.
    *   **ROC AUC (Receiver Operating Characteristic Area Under Curve)**: Measures the models' ability to discriminate between the positive and negative classes across all possible classification thresholds. An AUC score closer to 1 indicates better discriminatory power.

Calculate and plot the ROC AUC curve for both the MLP and Decision Tree models to evaluate their ability to discriminate between classes.

````python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Calculate ROC curve and AUC for MLP
y_pred_proba_mlp = mlp_model.predict_proba(X_test)[:, 1]
fpr_mlp, tpr_mlp, _ = roc_curve(y_test, y_pred_proba_mlp)
roc_auc_mlp = auc(fpr_mlp, tpr_mlp)

# Plot ROC curve for MLP
plt.figure()
plt.plot(fpr_mlp, tpr_mlp, color='darkorange', lw=2, label='MLP (AUC = %0.2f)' % roc_auc_mlp)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - MLP')
plt.legend(loc="lower right")
plt.show()

# Calculate ROC curve and AUC for Decision Tree
y_pred_proba_dt = dt_model.predict_proba(X_test)[:, 1]
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_pred_proba_dt)
roc_auc_dt = auc(fpr_dt, tpr_dt)

# Plot ROC curve for Decision Tree
plt.figure()
plt.plot(fpr_dt, tpr_dt, color='green', lw=2, label='Decision Tree (AUC = %0.2f)' % roc_auc_dt)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Decision Tree')
plt.legend(loc="lower right")
plt.show()
````

## Load and preprocess data

### Subtask:
Load the dataset from `/content/CloudWatch_Traffic_Web_Attack.csv`, handle missing values, and perform necessary feature engineering.


**Reasoning**:
Load the dataset into a pandas DataFrame and display the first few rows and check for missing values.



````python
import pandas as pd

df = pd.read_csv('/content/CloudWatch_Traffic_Web_Attack.csv')
print(df.head())
print(df.isnull().sum())
````

**Reasoning**:
There are no missing values in the dataset. Now, perform feature engineering by converting timestamp columns to datetime objects and extracting time-based features, and create a new feature for total bytes.



````python
df['creation_time'] = pd.to_datetime(df['creation_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['time'] = pd.to_datetime(df['time'])

df['duration'] = (df['end_time'] - df['creation_time']).dt.total_seconds()
df['hour'] = df['time'].dt.hour
df['dayofweek'] = df['time'].dt.dayofweek

df['total_bytes'] = df['bytes_in'] + df['bytes_out']

print(df[['duration', 'hour', 'dayofweek', 'total_bytes']].head())
````

## Model selection and training (black box)

### Subtask:
Choose and train a black-box model (e.g., a Neural Network) for threat detection.


**Reasoning**:
Define the target variable, prepare the data for the black-box model by selecting relevant features and splitting the data into training and testing sets, and encode categorical features.



````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np

# 1. Define the target variable
# Assuming 'detection_types' indicates threats. We'll treat any non-null/non-empty detection_type as a threat.
df['is_threat'] = df['detection_types'].apply(lambda x: 1 if pd.notna(x) and x != '' else 0)

# Introduce noise by randomly flipping a small percentage of labels
# Let's aim for flipping about 4% of the labels (this is an estimate and might require adjustment)
noise_level = 0.04
num_to_flip = int(len(df) * noise_level)
indices_to_flip = np.random.choice(df.index, num_to_flip, replace=False)

df.loc[indices_to_flip, 'is_threat'] = 1 - df.loc[indices_to_flip, 'is_threat']

# 2. Prepare the data
# Select features (excluding the original detection_types and time-related columns used for engineering)
features = ['bytes_in', 'bytes_out', 'duration', 'hour', 'dayofweek', 'total_bytes']
X = df[features]
y = df['is_threat']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# No categorical features in the selected set, so no encoding needed for these features.
# If we were to use 'src_ip' or 'dest_ip', we would need encoding.

print("Target variable 'is_threat' created with introduced noise.")
print(f"Original data shape: {df.shape}")
print(f"Training data shape (X_train, y_train): {X_train.shape}, {y_train.shape}")
print(f"Testing data shape (X_test, y_test): {X_test.shape}, {y_test.shape}")
print(f"Number of labels flipped: {num_to_flip}")
````

**Reasoning**:
Choose and train a black-box model suitable for the task, such as a Neural Network (e.g., a simple Multi-layer Perceptron).



````python
from sklearn.neural_network import MLPClassifier

# 3. Choose a black-box model (MLPClassifier)
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=300, random_state=42)

# 4. Train the chosen model on the training data
mlp_model.fit(X_train, y_train)

print("MLPClassifier model trained.")
````

**Reasoning**:
Evaluate the trained model's performance on the testing data using appropriate metrics for classification.



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# 5. Evaluate the trained model's performance
y_pred = mlp_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Model Evaluation:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Optional: Print a detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
````

````python
import streamlit as st
import pandas as pd
import joblib

# Define file paths for the saved models
MLP_MODEL_PATH = '/content/mlp_model.joblib'
DT_MODEL_PATH = '/content/dt_model.joblib'

# Use st.cache_resource to load the models once and cache them
@st.cache_resource
def load_models(mlp_path, dt_path):
    """Loads the trained MLP and Decision Tree models."""
    try:
        mlp_model = joblib.load(mlp_path)
        dt_model = joblib.load(dt_path)
        # st.success("Models loaded successfully!") # Removed success message from loader function
        return mlp_model, dt_model
    except FileNotFoundError:
        st.error(f"Error: Model file not found. Make sure '{mlp_path}' and '{dt_path}' are in the correct directory.")
        return None, None
    except Exception as e:
        st.error(f"An error occurred while loading the models: {e}")
        return None, None

# Load the models when the app runs
mlp_model, dt_model = load_models(MLP_MODEL_PATH, DT_MODEL_PATH)

# Assume X_train.columns is available from the training phase to ensure correct feature order
# In a real Streamlit app, you might save/load this separately or derive from a small sample of the training data
# For this example, we'll use a placeholder based on the features engineered earlier.
# In a full application, you should ensure this matches the columns your models were trained on.
# Let's add a placeholder for X_train.columns if it's not already in the session state or global scope
if 'X_train_columns' not in st.session_state:
    # This is a fallback. Ideally, X_train.columns should be saved during training and loaded here.
    # Assuming the engineered features are used:
    st.session_state['X_train_columns'] = ['bytes_in', 'bytes_out', 'duration', 'hour', 'dayofweek', 'total_bytes']

feature_columns = st.session_state['X_train_columns']


# Define the predict_threat function (updated to use st.session_state['X_train_columns'])
def predict_threat(instance_data: dict, mlp_model, dt_model, feature_columns_list):
    """
    Predicts threat using loaded MLP and Decision Tree models.

    Args:
        instance_data: Dictionary containing feature values for a single instance.
        mlp_model: Loaded MLP model.
        dt_model: Loaded Decision Tree model.
        feature_columns_list: List of feature names in the correct order used during training.

    Returns:
        A dictionary containing prediction results and explanations.
    """
    try:
        # Convert the input instance into a pandas DataFrame
        # Ensure the order of columns matches the training data
        instance_df = pd.DataFrame([instance_data], columns=feature_columns_list)

        # Ensure data types are appropriate (e.g., numeric)
        for col in feature_columns_list:
            if col in instance_df.columns:
                 # Use errors='coerce' to turn unparseable values into NaN
                 instance_df[col] = pd.to_numeric(instance_df[col], errors='coerce')

        # Handle potential NaNs introduced by coercion if necessary, e.g., fill with mean/median or drop row
        # For this simple app, we'll assume valid numeric input based on number_input widgets.
        # If using text input, more robust handling is needed.

        # Use the loaded models to get predictions
        mlp_prediction = mlp_model.predict(instance_df)[0]
        dt_prediction = dt_model.predict(instance_df)[0]

        # Optional: Get prediction probabilities for ROC AUC context or further analysis
        # Check if predict_proba is available for the models
        mlp_proba = mlp_model.predict_proba(instance_df)[:, 1][0] if hasattr(mlp_model, 'predict_proba') else "N/A"
        dt_proba = dt_model.predict_proba(instance_df)[:, 1][0] if hasattr(dt_model, 'predict_proba') else "N/A"


        # Determine if a threat is detected (Assuming class 1 is the threat class)
        is_threat_detected = False
        flagging_models = []

        if mlp_prediction == 1:
            is_threat_detected = True
            flagging_models.append("Black-box (MLP)")

        if dt_prediction == 1:
            is_threat_detected = True
            flagging_models.append("White-box (Decision Tree)")

        results = {
            "threat_detected": is_threat_detected,
            "flagging_models": flagging_models,
            "mlp_prediction": int(mlp_prediction), # Ensure serializable
            "dt_prediction": int(dt_prediction),   # Ensure serializable
            "mlp_proba": float(mlp_proba) if isinstance(mlp_proba, (int, float)) else mlp_proba, # Ensure serializable
            "dt_proba": float(dt_proba) if isinstance(dt_proba, (int, float)) else dt_proba,     # Ensure serializable
            "explanation_dt_placeholder": "Explanation based on Decision Tree rules would go here (or be generated by LLM).", # Placeholder
            "explanation_mlp_placeholder": "MLP is a black-box model, detailed explanation is difficult (LIME/SHAP could be used).", # Placeholder
            "mitigation_suggestion_placeholder": "Mitigation suggestion using Generative AI would go here." # Placeholder
        }

        return results

    except Exception as e:
        st.error(f"An error occurred during prediction: {e}")
        return None

# --- Streamlit App Layout and UI/UX Improvements ---

# 1. Clear introductory section
st.title('Cloud Network Threat Detection and Mitigation')
st.write("This application demonstrates an AI model for detecting potential cyber threats in cloud network traffic data. Enter the details of a network flow below to get a real-time threat assessment and potential mitigation suggestions.")
st.write("The system uses two models: a Black-box MLP (Neural Network) and a White-box Decision Tree, and conceptually integrates Generative AI for explanations and mitigation.")

# Add a sidebar for model info or future chatbot
with st.sidebar:
    st.subheader("About the Models")
    st.write("**Black-box Model (MLP):**")
    st.write("A Multi-layer Perceptron (Neural Network). It's powerful but its decision process is not easily interpretable.")
    st.write("**White-box Model (Decision Tree):**")
    st.write("A Decision Tree Classifier. Its rules are transparent, allowing us to understand why a prediction was made.")
    st.write("\nFuture integration will include a chatbot here for queries and detailed explanations.")

# Check if models are loaded before proceeding
if mlp_model is None or dt_model is None:
    st.warning("Models could not be loaded. Please check the file paths and try again.")
else:
    st.success("Models loaded successfully!") # Moved success message here

    st.subheader("Enter Network Flow Details")

    # 2. Organize input fields logically using columns
    col1, col2 = st.columns(2)

    # Default values for input fields based on the features engineered
    default_bytes_in = 5000
    default_bytes_out = 12000
    default_duration = 600.0
    default_hour = 9
    default_dayofweek = 4

    with col1:
        # 6. Include clear labels and helper text
        bytes_in = st.number_input('Bytes In', min_value=0, value=default_bytes_in, help="The number of bytes transferred in this flow towards the destination.")
        duration = st.number_input('Duration (seconds)', min_value=0.0, value=default_duration, help="The duration of the network flow in seconds.")
        dayofweek = st.number_input('Day of Week', min_value=0, max_value=6, value=default_dayofweek, help="The day of the week (0=Monday, 6=Sunday) when the flow occurred.")


    with col2:
        bytes_out = st.number_input('Bytes Out', min_value=0, value=default_bytes_out, help="The number of bytes transferred in this flow away from the destination.")
        hour = st.number_input('Hour of Day', min_value=0, max_value=23, value=default_hour, help="The hour of the day (0-23) when the flow occurred.")
        # Calculate total_bytes automatically
        total_bytes_calc = bytes_in + bytes_out
        # Display calculated total bytes, disabled for user input
        st.number_input('Total Bytes (calculated)', value=total_bytes_calc, disabled=True, help="The sum of Bytes In and Bytes Out.")


    # Prepare instance data from inputs, ensuring it matches expected feature columns
    instance_data_input = {
        'bytes_in': bytes_in,
        'bytes_out': bytes_out,
        'duration': duration,
        'hour': hour,
        'dayofweek': dayofweek,
        'total_bytes': total_bytes_calc # Use the calculated value
    }

    # Add a Streamlit button to trigger prediction
    if st.button('Detect Threat'):
        # Call the predict_threat function and display results
        prediction_results = predict_threat(instance_data_input, mlp_model, dt_model, feature_columns)

        if prediction_results:
            st.subheader("Threat Detection Results:")

            # 4. Incorporate visual cues (icons/colors) and 3. Enhance display
            if prediction_results["threat_detected"]:
                st.error(" **Potential Threat Detected!**") # Error box for threats
                st.metric(label="Threat Status", value="Threat Detected", delta_color="inverse") # Use metric for status
                st.write(f"**Flagged by:** {', '.join(prediction_results['flagging_models'])}")

                # Use expanders for detailed explanations and suggestions
                with st.expander("View Details and Explanations"):
                    st.write("---") # Separator
                    st.write("**Raw Model Predictions:**")
                    st.json({ # Display raw predictions as JSON
                        "MLP Prediction": prediction_results['mlp_prediction'],
                        "Decision Tree Prediction": prediction_results['dt_prediction']
                    })

                    # Display probabilities if available
                    st.write("**Prediction Probabilities (Threat Class):**")
                    st.json({ # Display probabilities as JSON
                        "MLP Probability": f"{prediction_results['mlp_proba']:.4f}" if isinstance(prediction_results['mlp_proba'], float) else prediction_results['mlp_proba'],
                        "Decision Tree Probability": f"{prediction_results['dt_proba']:.4f}" if isinstance(prediction_results['dt_proba'], float) else prediction_results['dt_proba']
                    })

                    st.write("**Explanation:**")
                    if "White-box (Decision Tree)" in prediction_results["flagging_models"]:
                         st.write("**Decision Tree:**")
                         st.write(prediction_results["explanation_dt_placeholder"])

                    if "Black-box (MLP)" in prediction_results["flagging_models"]:
                         st.write("**MLP:**")
                         st.write(prediction_results["explanation_mlp_placeholder"])
                    elif "White-box (Decision Tree)" not in prediction_results["flagging_models"]:
                         # If only MLP flagged it but DT didn't, still show MLP explanation placeholder
                         st.write("**MLP:**")
                         st.write(prediction_results["explanation_mlp_placeholder"])

                with st.expander("View Mitigation Suggestions"):
                    st.write("---") # Separator
                    st.write("**Mitigation Suggestion (using Generative AI):**")
                    st.write(prediction_results["mitigation_suggestion_placeholder"])


            else:
                st.success(" **Instance Classified as Non-Threat**") # Success box for non-threats
                st.metric(label="Threat Status", value="Non-Threat Detected", delta_color="normal") # Use metric for status
                st.write("No immediate action required.")

                with st.expander("View Details"):
                    st.write("---") # Separator
                    st.write("**Raw Model Predictions:**")
                    st.json({ # Display raw predictions as JSON
                        "MLP Prediction": prediction_results['mlp_prediction'],
                        "Decision Tree Prediction": prediction_results['dt_prediction']
                    })

                    # Display probabilities if available
                    st.write("**Prediction Probabilities (Threat Class):**")
                    st.json({ # Display probabilities as JSON
                        "MLP Probability": f"{prediction_results['mlp_proba']:.4f}" if isinstance(prediction_results['mlp_proba'], float) else prediction_results['mlp_proba'],
                        "Decision Tree Probability": f"{prediction_results['dt_proba']:.4f}" if isinstance(prediction_results['dt_proba'], float) else prediction_results['dt_proba']
                    })

            # Optional: Display the input data used for prediction
            # with st.expander("View Input Data"):
            #     st.dataframe(pd.DataFrame([instance_data_input])) # Display input data as dataframe
````
